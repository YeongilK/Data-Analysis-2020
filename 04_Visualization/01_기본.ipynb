# 데이터 시각화

### 시각화 패키지 맷플롯리브 소개

```python
# 주피터 노트북을 사용하는 경우에는 다음처럼 %matplotlib 매직(magic) 명령으로 노트북 내부에 그림을 표시하도록 지정해야 한다.
import matplotlib as mpl 
import matplotlib.pyplot as plt 
%matplotlib inline
```

### Line Plot

```python
plt.title('Line Plot')
plt.plot([1, 4, 9, 16])     # y좌표를 나타낸다 ( x좌표의 기본은 [0, 1, 2, 3] )
plt.show()
```

```python
plt.title('Line Plot')
plt.plot([1, 4, 9, 16])
plt.grid(True)
plt.show()
```

```python
plt.title('Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.plot([1, 4, 9, 16])
plt.grid(True)
plt.show()
```

```python
plt.title('x ticks')
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.grid(True)
plt.show()
```

### 한글폰트 사용

```python
import matplotlib.font_manager

matplotlib.font_manager._rebuild()
sorted([f.name for f in matplotlib.font_manager.fontManager.ttflist if f.name.startswith("Malgun")])
```

```python
# 폰트 설정
mpl.rc('font', family='Malgun Gothic')
# 유니코드에서  음수 부호설정
mpl.rc('axes', unicode_minus=False)
```

```python
plt.title('한글 제목')
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.xlabel("엑스축 라벨")
plt.ylabel("와이축 라벨")
plt.show()
```
