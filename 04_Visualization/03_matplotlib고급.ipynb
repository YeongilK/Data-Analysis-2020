#### Figure 객체

```python
np.random.seed(0)
f1 = plt.figure(figsize=(10, 2))
plt.title("figure size : (10, 2)")
plt.plot(np.random.randn(100))
plt.show()
```

```python
f1 = plt.figure(figsize=(15, 3))
plt.title("figure size : (15, 3)")
plt.plot(np.random.randn(100))
plt.show()
```

```python
f1 = plt.figure(1)
plt.title("현재의 Figure 객체")
plt.plot([1, 2, 3, 4], 'ro:')

f2 = plt.gcf()
print(f1, id(f1))
print(f2, id(f2))
plt.show()
```

```python
x1 = np.linspace(0.0, 5.0, 100)
x2 = np.linspace(0.0, 2.0, 100)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

```python
np.random.seed(2020)

plt.subplot(221)
plt.plot(np.random.rand(5))
plt.title("axes 1")

plt.subplot(222)
plt.plot(np.random.rand(5))
plt.title("axes 2")

plt.subplot(223)
plt.plot(np.random.rand(5))
plt.title("axes 3")

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title("axes 4")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2)

np.random.seed(2020)
axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title("axes 1")
axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")
axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title("axes 3")
axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")

plt.tight_layout()
plt.show()
```

```python
# for 반복문 사용
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)

np.random.seed(2020)
for i in range(2):
    for k in range(2):
        ax = axes[i][k]
        ax.plot(np.random.rand(5))
        ax.set_title(f'axes {2*i+k+1}')

plt.tight_layout()
plt.show()
```

```python
# for 반복문 사용
x = np.linspace(0.0, 5.0, 100)

fig, axes = plt.subplots(2, 1, sharex=True)
for i in range(2):
    ax = axes[i]
    y = np.cos(2 * np.pi * x1) * np.exp(-x1) if i == 0 else np.cos(2 * np.pi * x2)
    ylabel = "Damped oscillation" if i == 0 else "Undamped"
    ax.plot(x, y)
    ax.set_ylim(-1, 1)
    ax.set_ylabel(ylabel)

plt.suptitle('A tale of 2 subplots')
plt.xlabel('time (s)')
plt.show()
```

### Exercise - Iris Box Plot

```python
import seaborn as sns 
iris = sns.load_dataset('iris')
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 8), sharey=True)
species = ['setosa', 'versicolor', 'virginica']
for i, spec in enumerate(species):
    sl = iris[iris.species == spec].sepal_length
    sw = iris[iris.species == spec].sepal_width
    pl = iris[iris.species == spec].petal_length
    pw = iris[iris.species == spec].petal_width
    ax = axes[i]
    ax.boxplot((sl, sw, pl, pw), 
                labels=["Sepal length", "Sepal width", "Petal length", "Petal width"])
    ax.set_title(f'{spec} 품종')

plt.show()
```

### Exercise - Anscombe's Quartet

```python
ans = sns.load_dataset('anscombe')
ans.shape
```

```python
ans.head()
```

```python
ans.groupby('dataset').describe()
```

```python
ds1 = ans[ans['dataset'] == 'I']; ds2 = ans[ans['dataset'] == 'II']
ds3 = ans[ans['dataset'] == 'III']; ds4 = ans[ans['dataset'] == 'IV']
```

```python
fig = plt.figure(figsize=(8, 10))
ax1 = plt.subplot(2, 2, 1); ax2 = plt.subplot(2, 2, 2)
ax3 = plt.subplot(2, 2, 3); ax4 = plt.subplot(2, 2, 4)

ax1.plot(ds1['x'], ds1['y'], 'o'); ax2.plot(ds2['x'], ds2['y'], 'o')
ax3.plot(ds3['x'], ds3['y'], 'o'); ax4.plot(ds4['x'], ds4['y'], 'o')

ax1.set_title('Dataset I'); ax2.set_title('Dataset II')
ax3.set_title('Dataset III'); ax4.set_title('Dataset IV')

fig.suptitle("Anscombe's Quartet", fontsize=20, y=1.05)
fig.tight_layout()
```

#### Code Refactoring

```python
ds1 = ans[ans['dataset'] == 'I']; ds2 = ans[ans['dataset'] == 'II']
ds3 = ans[ans['dataset'] == 'III']; ds4 = ans[ans['dataset'] == 'IV']

fig, axes = plt.subplots(1, 4, figsize=(10, 8), sharey=True)
set = [ds1, ds2, ds3, ds4]
for i, ds in enumerate(set):
    ax = axes[i]
    ax.plot(ds['x'], ds['y'], 'o')
    ax.set_title(f'Dataset {i+1}')

plt.suptitle("Anscombe's Quartet", fontsize=20, y=1.05)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8), sharex=True, sharey=True)
for i, num in enumerate(['I', 'II', 'III', 'IV']):
    ds = ans[ans.dataset == num]
    ax = axes[i//2][i%2]
    ax.plot(ds.x, ds.y, 'o', markersize=10)
    ax.plot([3, 20], [4.5, 13], 'r-', linewidth=2)
    ax.set_title(f'Dataset {num}')

plt.suptitle("Anscombe's Quartet", fontsize=20, y=1.05)
plt.tight_layout()
```

#### Axis 객체와 축

```python
fig, ax0 = plt.subplots()
ax1 = ax0.twinx()
ax0.set_title("2개의 y축 한 figure에서 사용하기")
ax0.plot([10, 5, 2, 9, 7], 'r-', label="y0")
ax0.set_ylabel("y0")
ax0.grid(False)
ax1.plot([100, 200, 220, 180, 120], 'g:', label="y1")
ax1.set_ylabel("y1")
ax1.grid(False)
ax0.set_xlabel("공유되는 x축")
plt.legend()
plt.show()
```
