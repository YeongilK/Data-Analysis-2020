# Exercise - 강서구 커피 매장 정보

```python
import pandas as pd 
import numpy as np
import folium
```

```python
df = pd.read_csv('../input/coffee.csv', encoding='euc-kr')
df.head()
```

```python
len(df.index)
```

```python
# 도로명주소에 '서울특별시 강서구' 를 포함한 행만 추출
keyword = '서울특별시 강서구'
df2 = df.loc[df['도로명주소'].str.contains(keyword, na=False)]
len(df2)
```

#### 서울특별시 강서구에 있는 스타벅스, 이디야, 빽다방, 투썸플레이스 매장만 추출하여 데이터프레임에 저장

```python
starbucks = df2.loc[df['상호명'].str.contains('스타벅스', na=False)]
ediya = df2.loc[df['상호명'].str.contains('이디야', na=False)]
paikdabang = df2.loc[df['상호명'].str.contains('빽다방', na=False)]
twosome = df2.loc[df['상호명'].str.contains('투썸', na=False)]
```

```python
starbucks['color'] = 'cadetblue'; starbucks['icon'] = 'usd'
ediya['color'] = 'red'; ediya['icon'] = 'tag'
paikdabang['color'] = 'purple'; paikdabang['icon'] = 'map-marker'
twosome['color'] = 'orange'; twosome['icon'] = 'certificate'
```

```python
tmp1 = pd.concat([starbucks, ediya])
tmp2 = pd.concat([paikdabang, twosome])
df3 = pd.concat([tmp1, tmp2], ignore_index=True)
df3
```

```python
# 새로운 csv 파일로 저장
df3.to_csv('../input/coffee2.csv', sep=',', encoding='utf8', index=False)
```

### folium을 사용하여 지도 그리기

#### 스타벅스, 이디야, 빽다방, 투썸플레이스 전체 매장

```python
# 매장들의 위도와 경도의 평균을 중심으로 한 배경 지도
mapping = folium.Map(location=[df3['위도'].mean(), df3['경도'].mean()], zoom_start=13)

for i in df3.index:
    folium.Marker(
        location=[df3['위도'][i], df3['경도'][i]],
        popup=df3['도로명주소'][i], 
        tooltip=f"{df3['상호명'][i]}, {df3['지점명'][i]}",
        icon=folium.Icon(color=df3.color[i], icon=df3.icon[i])
    ).add_to(mapping)
mapping
```

#### 스타벅스 매장

```python
starbucks.reset_index(drop=True, inplace=True)

# 스타벅스 매장들의 위도와 경도의 평균을 중심으로 한 배경 지도
mapping = folium.Map(location=[starbucks['위도'].mean(), starbucks['경도'].mean()], zoom_start=13)

for i in starbucks.index:
    folium.Marker(
        location=[starbucks['위도'][i], starbucks['경도'][i]],
        popup=starbucks['도로명주소'][i], 
        tooltip=f"{starbucks['상호명'][i]}, {starbucks['지점명'][i]}",
        icon=folium.Icon(color='cadetblue', icon='usd')
    ).add_to(mapping)
mapping
```

```python
starbucks.to_csv('../input/starbucks.csv', sep=',', encoding='utf8', index=False)
```
